```python
# Dependencies and Setup
import os
import numpy as np
import pandas as pd
from pathlib import Path

#declare variables required
total_number_schools = 0
Total_number_of_students = 0

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Initialize the list to be used to build the dataframe of results
results_list =[{'Total Schools':0, 'Total Students':0, 'Total Budget': 0.0, 'Average Maths Score': 0.0, 'Average Reading Score':0.0, '% Passing Maths':0.0, '% Passing Reading':0.0, '% Overall Passing':0.0}]
school_results_df = pd.DataFrame(results_list)
school_results_df.head()
#Calculate the total number of schools
total_number_of_schools = len(school_data)
school_results_df['Total Schools'] = total_number_of_schools
school_results_df.head()
#Calculate the total number of students
total_number_of_students = len(student_data)
school_results_df['Total Students'] = total_number_of_students
school_results_df.head()
#Calculate the total budget
total_schools_budget = school_data['budget'].sum(axis=0)
school_results_df['Total Budget'] = total_schools_budget
school_results_df['Total Budget'] = school_results_df['Total Budget'].map('${:,d}'.format)
# learnt how to format numbers df.loc[:, "Population"] = df["Population"].map('{:,d}'.format)
# learnt how to format numbers df["Value"] = df["Value"].map("${:,.0f}K".format)
#Calculate the Average Maths Score
average_maths_score = student_data['maths_score'].mean(axis=0)
school_results_df['Average Maths Score'] = average_maths_score

#Calculate the Average Reading Score
average_reading_score = student_data['reading_score'].mean(axis=0)
school_results_df['Average Reading Score'] = average_reading_score
#Calculate % Students who passed maths
students_passed_maths_df = student_data.loc[(student_data['maths_score'] >= 50)]
#print(students_passed_maths_df)
school_results_df['% Passing Maths'] = (len(students_passed_maths_df)/total_number_of_students)*100
#Calculate % Students who passed reading
students_passed_reading_df = student_data.loc[(student_data['reading_score'] >= 50)]
school_results_df['% Passing Reading'] = (len(students_passed_reading_df)/total_number_of_students)*100
#Calculate % Students who passed overall
students_passed_overall_df = student_data.loc[(student_data['maths_score'] >= 50) & (student_data['reading_score'] >= 50)]

school_results_df['% Overall Passing'] = (len(students_passed_overall_df)/total_number_of_students)*100
#school_results_df['% Passing Maths'] = school_results_df['% Passing Maths'].map('${:,d}'.format)

school_results_df
#print("Total number of schools is: " + str(total_number_of_schools))

#calculate the total number of students
```

```python
#Group the merged dataframe by school to generate the basic data
school_data_complete_group_df = school_data_complete.groupby(['school_name','type'], as_index=False).agg({'Student ID':'count','reading_score': 'mean', 'maths_score':'mean','budget':'mean'})
school_data_complete_group_df= pd.DataFrame(school_data_complete_group_df)
school_data_complete_group_df
```

```python
# find the number of students who passed maths in each school
# sort the school data
i = 0
school_data_overview_list =[]
#school_data_passing_math_df = pd.DataFrame(school_data_passing_math_list)
#school_data_passing_math_df
print("No of schools: " + str(len(school_data_complete_group_df)))
for i in range(len(school_data_complete_group_df)):
    #print(str(i))
    current_school = school_data_complete_group_df.iloc[i,0]
    current_school_type = school_data_complete_group_df.iloc[i,1]
    current_school_student_count = school_data_complete_group_df.iloc[i,2]
    current_school_total_budget = school_data_complete_group_df.iloc[i,5]
    current_school_per_student_budget = current_school_total_budget / current_school_student_count 
    current_school_student_average_maths_score = school_data_complete_group_df.iloc[i,4]
    current_school_student_average_reading_score = school_data_complete_group_df.iloc[i,3]
    #print("current school is " + current_school)
    filter_students_passed_maths_df = school_data_complete.loc[(school_data_complete['maths_score'] >= 50) & (school_data_complete['school_name'] == current_school)]
    filter_students_passed_reading_df = school_data_complete.loc[(school_data_complete['reading_score'] >= 50) & (school_data_complete['school_name'] == current_school)]
    filter_students_passed_overall_df = school_data_complete.loc[(school_data_complete['maths_score'] >= 50)  & (school_data_complete['reading_score'] >= 50) & (school_data_complete['school_name'] == current_school)]
    no_passed_maths = len(filter_students_passed_maths_df)
    current_school_percent_passed_maths = (no_passed_maths/current_school_student_count)*100
    no_passed_reading = len(filter_students_passed_reading_df)
    current_school_percent_passed_reading = ((no_passed_reading)/current_school_student_count)*100
    no_passed_overall = len(filter_students_passed_overall_df)
    current_school_percent_passed_overall = ((no_passed_overall)/current_school_student_count)*100
    #school_results_df['Total Budget'] = school_results_df['Total Budget'].map('${:,d}'.format)
    #write data into the list
    school_data_overview_list.append( {'School': current_school,
                                        'School Type': current_school_type,
                                        'Total Students': current_school_student_count,
                                        'Total School Budget': current_school_total_budget,
                                        'Per Student Budget': current_school_per_student_budget,
                                        'Average Maths Score' : current_school_student_average_maths_score,
                                        'Average Reading Score': current_school_student_average_reading_score,
                                        '% Passing Maths': current_school_percent_passed_maths,
                                        '% Passing Reading': current_school_student_average_reading_score,
                                        '% Overall Passing': current_school_percent_passed_overall})
school_data_overview_df = pd.DataFrame(school_data_overview_list)
#school_data_overview_df['Total School Budget'] = school_data_overview_df['Total School Budget'].map('${:,d}'.format)
#school_data_overview_df['Per Student Budget'] = school_data_overview_df['Per Student Budget'].map('${:,d}'.format)
school_data_overview_df

```
